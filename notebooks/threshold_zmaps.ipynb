```python
import os.path as op
import os
from nilearn.image import threshold_img
```

```python
zmaps_path = '/scratch/hyruuk/neuromod/shinobi2023/processed/z_maps'

for level in os.listdir(zmaps_path):
    level_path = op.join(zmaps_path, level)
    for cond in os.listdir(level_path):
        cond_path = op.join(level_path, cond)
        for zmap in os.listdir(cond_path):
            print(zmap)
            zmap_path = op.join(cond_path, zmap)
            thresholded_zmap = threshold_img(zmap_path, threshold=3)
            os.makedirs(cond_path.replace('z_maps', 'thresholded_z_maps'), exist_ok=True)
            thresholded_zmap.to_filename(zmap_path.replace('z_maps', 'thresholded_z_maps').replace('.nii.gz', '.nii'))
```
